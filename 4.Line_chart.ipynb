```python
import pandas as pd  
import matplotlib.pyplot as plt  
```

```python
df_line = pd.read_csv("line_chart_data.csv")
df_line
```

Change the date columns in the data set to a date format.

```python
df_line['New_date'] = pd.to_datetime(df_line['Date'])
df_line
```

```python
labels = ['S&P 500','FISE 100']
plt.figure(figsize=(20,8) , dpi = 500)
plt.plot(df_line['New_date'],df_line['GSPC500'])
plt.plot(df_line['New_date'],df_line['FTSE100'])
plt.title("S&P vs FTSE Return (2000-2010)",fontsize=14,fontweight = 'bold')
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels=labels,fontsize = 'large')

plt.show()
```
