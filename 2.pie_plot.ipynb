```python
import pandas as pd     
import matplotlib.pyplot as plt  
import seaborn as sns   
sns.set()
```

```python
data = pd.read_csv('pie_chart_data.csv')
data
```

```python
plt.figure(figsize=(10, 10))
plt.pie(data['Number of Cars'] , labels = data['Engine Fuel Type'].values,autopct='%.2f%%' ,textprops={'size':'large', 'fontweight':'bold', 'rotation': 30,'color':'w'})
plt.legend()
plt.title('Cars by Engine Fuel Type', fontsize=20, fontweight='bold')
plt.show()
```
