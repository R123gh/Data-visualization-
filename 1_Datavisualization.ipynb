```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20] 

plt.figure(figsize=(5, 3))

plt.plot(x, y, label='Line', color='blue', linestyle='--', marker='o')
plt.title('Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True)
plt.show()
```

# Bar Plot

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.figure(figsize=(5,3))

plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

 #  Horizontal Bar Plot

```python
plt.barh(categories, values, color='lightgreen')
plt.title('Horizontal Bar Chart')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
plt.figure(figsize=(5,3))
```

#  Scatter Plot

```python
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 6, 7] 

plt.figure(figsize=(5,3)) 

plt.scatter(x, y  , color = 'red')
plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()

```

#  Histogram

Used for distribution of numerical data

```python
data = np.random.randn(1000)

plt.figure(figsize=(5,3))

plt.hist(data, bins = 30 , color = 'purple',edgecolor = 'black' )
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

#  Pie Chart

```python
labels = ['Python','C','Javascript','React-Native']
size = [45,30,15,10]
color = ['gold','lightblue','lightcoral','pink']


plt.pie(size, labels=labels, colors=color, autopct='%1.1f%%', startangle=140)
plt.title('Pie chart')
plt.axis('equal')
plt.legend(loc='lower right')
plt.show()
```

#  Subplots (Multiple Plots in One Figure)

```python
x = [1, 2, 3, 4, 5]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]
plt.figure(figsize=(5, 3))
plt.subplot(1, 2, 1)  
plt.plot(x, y1, color='green')
plt.title('Square Function')

plt.subplot(1, 2, 2)
plt.plot(x, y2, color='orange')
plt.title('Cube Function')

plt.tight_layout()
plt.show()
```

# Customizing Plots

plt.title(): Adds title

plt.xlabel(), plt.ylabel(): Axis labels

plt.legend(): Adds legend

plt.grid(True): Adds grid

plt.xlim() / plt.ylim(): Set axis limits

plt.xticks() / plt.yticks(): Customize ticks

```python
plt.figure(figsize=(5,3))
plt.style.use('ggplot')
plt.plot(x, y1, label='x^2', linestyle='-', marker='s', color='blue')
plt.plot(x, y2, label='x^3', linestyle='--', marker='^', color='red')
plt.title("Functions of x")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid(True)
plt.xlim(0, 6)
plt.ylim(0, 150)
plt.show()
```

```python
plt.plot(x, y1)
plt.title("Saving Plot")
plt.savefig("myplot.png", dpi=300) 
plt.show()
```

#  3D Plot (Advanced)

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = [1, 2, 3, 4, 5]
y = [5, 6, 2, 3, 13]
z = [2, 3, 3, 3, 5]

ax.scatter(x, y, z)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.title("3D Scatter Plot")
plt.show()
```

#  Animation (Bonus)

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()
x_data, y_data = [], []
line, = ax.plot([], [], 'r-', label='y = x^2')

def init():
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 100)
    line.set_data([], [])
    return line,

def animate(i):
    x_data.append(i)
    y_data.append(i**2)
    line.set_data(x_data, y_data)
    return line,

ani = animation.FuncAnimation(fig, animate, init_func=init,
                              frames=10, interval=500, blit=True)

plt.title("Animation Example: y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
