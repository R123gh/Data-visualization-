```python
import seaborn as sns 
import matplotlib.pyplot as plt    
```

Directly load the data using seaborn lib

```python
df = sns.load_dataset('tips')
df
```

Lets make distribution of the data using the histplot function of seaborn.


```python
plt.figure(figsize=(8,5), dpi=500)
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins= 8 , kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins= 8 , kde = True  , color= 'pink')
plt.show()
```

How to implement joinplot in Seaborn

```python
plt.figure(figsize=(8,5))
sns.jointplot(x = 'total_bill', y = 'tip' , data = df , kind = 'reg' )
plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = df , kind = 'hex' )
```

How to implememt the pairplot in seaborn

```python
sns.pairplot(df , hue = 'sex')
plt.figure(figsize=(8,5),)
```

```python
sns.pairplot(df , hue = 'size' , palette = 'rainbow')
```

Rugplot 

```python
sns.rugplot(df)
```

# How to implement the countplot in seaborn?

```python
sns.countplot(x = df['sex'])
```

```python
sns.countplot(x = df['sex']  , hue=df['smoker']  ,color='pink')
```

```python
sns.barplot(x = df['sex'], y = df['total_bill']  , color = '#AEC8A4'  )
```

```python
import numpy as np 
sns.barplot(x = df['sex'], y = df['total_bill'] , estimator=np.sum , color = '#0065F8'  )
```
