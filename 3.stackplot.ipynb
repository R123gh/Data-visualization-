```python
import pandas as pd  
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('stacked_area_chart_data.csv')
data
```

```python
colors = ['#D5451B','#FF9B45','#521C0D']
labels = ['Diesel','Petrol','Gas']
plt.figure(figsize=(10,6),dpi = 500)
plt.stackplot(data['Year'], data['Diesel'], data['Petrol'], data['Gas'], labels=labels, colors=colors)
plt.xticks(data['Year'], rotation=45)
plt.legend(loc='upper left')
plt.title('Stacked Area Chart of Fuel Consumption Over Years')
plt.xlabel('Year')
plt.ylabel('Fuel Consumption (Liters)')
plt.show()





```
