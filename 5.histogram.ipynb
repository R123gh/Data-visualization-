```python
import pandas as pd   
import matplotlib.pyplot as plt  
import seaborn as sns 
sns.set()
```

```python
scatter_data = pd.read_csv('scatter_data.csv')
```

```python
scatter_data
```

```python
sns.set_style("darkgrid")
plt.figure(figsize = (8,5) , dpi = 500)
plt.hist(scatter_data['Price'],bins = 8  , color = '#A53860' , edgecolor = 'black' , alpha = 0.5 , label = 'Price' )
plt.title('Histogram of Price'  , fontsize = 14 , fontweight = 'bold')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend()
sns.despine()
plt.show()
```
