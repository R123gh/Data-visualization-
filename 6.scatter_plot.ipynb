```python
import pandas as pd  
import matplotlib.pyplot as plt  
```

```python
scatter_data = pd.read_csv('scatter_data.csv')
scatter_data
```

```python
plt.figure(figsize = (8,5) , dpi = 500)
colors = ['#B2D8CE','#648DB3','#5459AC','#52357B']
plt.scatter(x=scatter_data['Area (ft.)'], y=scatter_data['Price'], color=colors[3], alpha=0.3)
plt.xlabel('Area (ft.)')
plt.ylabel('Price ($)')
plt.title('Scatter Plot of Area vs Price')
plt.show()
```
